```python
df=pd.read_csv('BankChurners.csv', index_col=0)
```

```python
df.head(2)
```

```python
#deleting last two rows
```

```python
df = df.iloc[:, :-2]
```

```python
df.head(0) #checking
```

```python
#rename columns
```

```python
df.rename(columns={"Customer_Age": "Age",
    "Dependent_count": "Dependents",
    "Months_on_book": "Tenure_Months",
    "Months_Inactive_12_mon": "Inactive_Months",
    "Contacts_Count_12_mon": "Contacts_Last_12_Months",
    "Total_Trans_Amt": "Total_Transaction_Amount",
    "Total_Trans_Ct": "Total_Transaction_Count"
}, inplace=True)
```

```python
df.rename(columns={"Attrition_Flag":"Churn"}, inplace=True)
```

```python
print(df.columns)
```

```python
#assigning existing customer to 1 and attrited to 0
```

```python
df["Churn"] = df["Churn"].map({"Existing Customer": 0,"Attrited Customer": 1})
```
